# LOAN ELIGIBILITY PREDICTION

```python
import pandas as pd

df = pd.DataFrame({
    "Gender": ["Male", "Female", "Male", "Male", "Female", "Female", "Male", "Female"],
    "Married": ["Yes", "No", "Yes", "Yes", "No", "No", "Yes", "No"],
    "Education": ["Graduate", "Not Graduate", "Graduate", "Graduate", "Not Graduate", "Graduate", "Graduate", "Graduate"],
    "Self_Employed": ["No", "Yes", "No", "No", "No", "Yes", "No", "No"],
    "ApplicantIncome": [5000, 3000, 4000, 6000, 2500, 3500, 7000, 2800],
    "LoanAmount": [150, 80, 120, 200, 70, 100, 250, 60],
    "Loan_Status": [1, 0, 1, 1, 0, 0, 1, 0]
})

df
```

```python
X = df.drop("Loan_Status", axis=1)
y = df["Loan_Status"]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_cols = ["Gender", "Married", "Education", "Self_Employed"]

encoder = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(drop="first"), categorical_cols)
    ],
    remainder="passthrough"
)

X_encoded = encoder.fit_transform(X)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_encoded, y)
```

```python
new_applicant = pd.DataFrame({
    "Gender": ["Male"],
    "Married": ["Yes"],
    "Education": ["Graduate"],
    "Self_Employed": ["No"],
    "ApplicantIncome": [4500],
    "LoanAmount": [120]
})

new_encoded = encoder.transform(new_applicant)

prediction = model.predict(new_encoded)

print("Loan Approved?" , "Yes" if prediction[0] == 1 else "No")
```
